```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Luxury_Housing_Bangalore.csv")
```

```python
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()
```

```python
df.head()
```

```python
df['micro_market'].value_counts()
```

```python
df['micro_market'] = df['micro_market'].str.title()
```

```python
df['micro_market'].value_counts()
```

```python
df['developer_name'].value_counts()
```

```python
df.info()
```

```python
df['configuration'] = df['configuration'].str.upper()
```

```python
df['configuration'].value_counts()
```

```python
df.head(10)
```

```python
df['ticket_price_cr'] = df['ticket_price_cr'].str.replace('₹','').str.replace('Cr','').str.strip()
```

```python
df.head(10)
```

```python
df['ticket_price_cr'] = pd.to_numeric(df['ticket_price_cr'])
```

```python
df.info()
```

```python
df.tail(10)
```

```python
df['nri_buyer'].value_counts()
```

```python
df['purchase_quarter'] = pd.to_datetime(df['purchase_quarter'])
```

```python
df.info()
```

```python
df['year'] = df['purchase_quarter'].dt.year
df['quarter'] = df['purchase_quarter'].dt.to_period('Q').astype(str)
```

```python
df.head()
```

```python
df.duplicated()
```

```python
df[df['property_id']=='PROP093022']
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['bhk_count']=df['configuration'].str[0].astype(int)
```

```python
df['bhk_count'].value_counts()
```

```python
df.info()
```

```python
df[df['unit_size_sqft']==-1]
```

```python
df['unit_size_sqft'] = df['unit_size_sqft'].replace(-1, np.nan)
```

```python
df.info()
```

```python
df['unit_size_sqft'] = (
    df.groupby('bhk_count')['unit_size_sqft'].transform(lambda x: x.fillna(x.median()))
)
```

```python
df.info()
```

```python
df['connectivity_score'].fillna(df['connectivity_score'].median(), inplace=True)
df['amenity_score'].fillna(df['amenity_score'].median(), inplace=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['ticket_price_cr'] = (
    df.groupby(['micro_market', 'bhk_count'])['ticket_price_cr'].transform(lambda x: x.fillna(x.median()))
)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
query="""SELECT count(*) AS Total_Rows FROM sales LIMIT 10"""
df1=pd.read_sql(query,engine)
df1
```

```python
query="""SELECT possession_status, count(*) as No_of_properties FROM sales GROUP BY possession_status"""
df1=pd.read_sql(query,engine)
df1
```

```python
query="""SELECT developer_name, avg(ticket_price_cr) as Avg_price_cr, count(*) as bookings_count FROM sales GROUP BY developer_name"""
df1=pd.read_sql(query,engine)
df1
```

```python
query="""SELECT micro_market, avg(ticket_price_cr) as Avg_price_cr FROM sales GROUP BY micro_market ORDER BY Avg_price_cr DESC """
df1=pd.read_sql(query,engine)
df1
```

```python
query="""SELECT quarter, count(*) as booking_count FROM sales GROUP BY quarter ORDER BY quarter """
df1=pd.read_sql(query,engine)
df1
```
